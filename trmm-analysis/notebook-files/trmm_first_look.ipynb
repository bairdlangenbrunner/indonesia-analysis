```python
import xarray
import matplotlib.pyplot as mp
import cartopy
```

```python
filename = '/Users/baird/Downloads/3B42_Daily.20180729.7.nc4'
```

```python
trmm_ds = xarray.open_dataset(filename)
```

```python
trmm_ds
```

```python
trmm_lat = trmm_ds['lat'].values
trmm_lon = trmm_ds['lon'].values
trmm_precip = trmm_ds['precipitation'].values
```

```python
data_proj = cartopy.crs.PlateCarree()
map_proj = cartopy.crs.PlateCarree()

fig = mp.figure(figsize=(8.5,6))

ax = fig.add_subplot(111, projection=map_proj)
plot = ax.pcolormesh(trmm_lon,trmm_lat,trmm_precip.T)
ax.coastlines(color='firebrick')

ax.set_extent([90,130,-10,10])
mp.colorbar(plot)
```
