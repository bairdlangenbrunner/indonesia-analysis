````python
filename = '../modis_data/MYD11A1.A2018174_data/MYD11A1.A2018174.h28v11.006.2018184003341.hdf'
````

open as xarray dataset first

* I find it's easier to grab the data themself using ```xarray```, and then used ```gdal``` to get info about the HDF file's projection

````python
modis_xarray_ds = xarray.open_dataset(filename)

lst = modis_xarray_ds['LST_Day_1km']
qcday = modis_xarray_ds['QC_Day']
dayviewtime = modis_xarray_ds['Day_view_time']
````

this is different from LandSAT because in the MODIS files there are "subdatasets" (each corresponding to a field of interest), and you have to extract the specific one you want using the ```[0][0]``` syntax below

````python
gdal_file = gdal.Open(filename)
all_modis_ds = gdal_file.GetSubDatasets()
modis_ds = gdal.Open(all_modis_ds[0][0])
modis_geotrans = modis_ds.GetGeoTransform()
````

use this to get info on the projection itself

* note that it's a "custom spheroid," which is import when actually plotting the MODIS data
* below, a ```modis_globe``` object is created, and the axis ```globe=``` keyword is used
* I tried using either an ellipse or a sphere, and only the sphere worked; the value below gives the semimajor axis of the MODIS projection sphere

````python
modis_proj_info = modis_ds.GetProjection()
modis_sro = osr.SpatialReference()
modis_sro.ImportFromWkt(modis_proj_info)
print(modis_sro)
````

the geotransform gives info on pixel size and beginning/end of data

````python
print(modis_geotrans)
````

````python
data_extent = (modis_geotrans[0], \
               modis_geotrans[0] + modis_ds.RasterXSize*modis_geotrans[1], \
               modis_geotrans[3] + modis_ds.RasterYSize*modis_geotrans[5], \
               modis_geotrans[3])
````

plot a very quick image, make sure it looks reasonable before you overlay coastlines

````python
mp.imshow(lst.values, origin='upper')
````

first overlay the coastlines on the native grid

* note you have to create your own ```globe``` object that tells cartopy what precise projection the MODIS data come in (it's irregular, and by default cartopy doesn't get it correct)
* you know it's working if the coastlines are flush with the actual data below... took a while to get this correct!

````python
modis_globe = cartopy.crs.Globe(datum='WGS84', ellipse='sphere', semimajor_axis=6371007.181)
map_proj = cartopy.crs.Sinusoidal(central_longitude=0, false_northing=0, false_easting=0, globe=modis_globe)
data_proj = cartopy.crs.Sinusoidal(central_longitude=0, false_northing=0, false_easting=0, globe=modis_globe)
geodetic_proj = cartopy.crs.Geodetic()

fig = mp.figure(figsize=(8.5,7))
ax = fig.add_subplot(111, projection=map_proj)
image = ax.imshow(lst.values, extent=(data_extent[0],data_extent[1],data_extent[2],data_extent[3]), origin='upper', transform=data_proj)
ax.coastlines(resolution='10m')

mp.colorbar(image)
````

now put this in a projection you want

* note this is typically a VERY SLOW process with default cartopy (took ~4 hours on my laptop using the default method)
* I had to go into the cartoy source code and change a line for the projection to be quicker, and it takes about 30s now...
* what I changed is below, located (on my laptop) in ```Users/baird/anaconda/lib/python3.6/site-packages/cartopy/img_transform.py```

**edit the line below**
```
try:
    kdtree = scipy.spatial.cKDTree(xyz, balanced_tree=False)
```
**to**
```
try:
    kdtree = scipy.spatial.cKDTree(xyz, balanced_tree=False, compact_nodes=False)
```

* see the [stackoverflow question here](https://stackoverflow.com/questions/31819778/scipy-spatial-ckdtree-running-slowly) for more info on this
* in the figure below, you can see that the reprojection looks a bit worse than the origina data, meaning there's potentially some information loss going on when plotting in a non-native projection, so it's probably a good idea to do most analysis on the MODIS native grid and then project later
