```python
map_proj = cartopy.crs.PlateCarree()
data_proj = cartopy.crs.PlateCarree()

fig = mp.figure(figsize=(8.5,8.5))

ax = fig.add_subplot(111, projection=map_proj)
ax.set_extent([95,120,-10,10])
ax.coastlines(resolution='50m')

lat_vals = site_info_indo['latitude'].values
lon_vals = site_info_indo['longitude'].values

ax.scatter(lon_vals,lat_vals,facecolor='firebrick',edgecolor='0.1',s=100,marker='^',zorder=2)
for row in range(site_info_indo.shape[0]):
    year_began = site_info_indo['year_began'].values[row]
    ax.text(s=year_began, x=lon_vals[row]-0.25, y=lat_vals[row], ha='right', va='center', fontsize=12)
    

fig.tight_layout()
fig.savefig('../figures/fluxnet_sites_indonesia_and_malaysia.pdf', bbox_inches='tight', transparent=True)
```
